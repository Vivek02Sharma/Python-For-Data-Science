## Feature Engineering

```python
df.drop(df[['Club_Join_Date']],axis = 1,inplace = True)
```

```python
df.head()
```

## 1.Handling missing value

```python
df.isnull().sum()
```

*Conclusion -> There is no missing value*

## 2.Handling outliers 

```python
# Predefined colors for each boxplot
colors = ['#FF6347', '#4682B4', '#32CD32', '#FFD700']

# Creating subplots
fig, axs = plt.subplots(2, 2, figsize=(25, 25))

sns.boxplot(df['Math_Score'], ax=axs[0, 0], color=colors[0])
axs[0, 0].set_title("Math Score")

sns.boxplot(df['Reading_Score'], ax=axs[0, 1], color=colors[1])
axs[0, 1].set_title("Reading Score")

sns.boxplot(df['Writing_Score'], ax=axs[1, 0], color=colors[2])
axs[1, 0].set_title("Writing Score")

sns.boxplot(df['Placement_Score'], ax=axs[1, 1], color=colors[3])
axs[1, 1].set_title("Placement Score")

plt.show()
```

*Conclusion -> As we can see there is no outliers in graph.So we don't have to handle outliers*

## 3.Feature Scaling

```python
df.head()
```

*Before scaling the graph*

```python
# Predefined colors for each boxplot
colors = ['#FF6347', '#4682B4', '#32CD32', '#FFD700']

# Creating subplots
fig, axs = plt.subplots(2, 2, figsize=(25, 25))

sns.histplot(df['Math_Score'], kde = True, ax=axs[0, 0], color=colors[3])
axs[0, 0].set_title("Math Score")

sns.histplot(df['Reading_Score'], kde = True, ax=axs[0, 1], color=colors[2])
axs[0, 1].set_title("Reading Score")

sns.histplot(df['Writing_Score'], kde = True, ax=axs[1, 0], color=colors[1])
axs[1, 0].set_title("Writing Score")

sns.histplot(df['Placement_Score'], kde = True, ax=axs[1, 1], color=colors[0])
axs[1, 1].set_title("Placement Score")

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard_scaler = StandardScaler()
```

```python
scaled_data = standard_scaler.fit_transform(df)
```

```python
df1 = pd.DataFrame(scaled_data,columns = df.columns.to_list())
```

```python
df1.head()
```

*After scaling the graph*

```python
# Predefined colors for each boxplot
colors = ['#FF6347', '#4682B4', '#32CD32', '#FFD700']

# Creating subplots
fig, axs = plt.subplots(2, 2, figsize=(25, 25))

sns.histplot(df1['Math_Score'], kde = True, ax=axs[0, 0], color=colors[3])
axs[0, 0].set_title("Math Score")

sns.histplot(df1['Reading_Score'], kde = True, ax=axs[0, 1], color=colors[2])
axs[0, 1].set_title("Reading Score")

sns.histplot(df1['Writing_Score'], kde = True, ax=axs[1, 0], color=colors[1])
axs[1, 0].set_title("Writing Score")

sns.histplot(df1['Placement_Score'], kde = True, ax=axs[1, 1], color=colors[0])
axs[1, 1].set_title("Placement Score")

plt.show()
```

*Conclusion -> As you can see ,there is no difference in after and before graph.*
*That means data is properly scaled*
