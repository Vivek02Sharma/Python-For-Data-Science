## Min-max scaling (Normalization)

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('healthexp')
```

```python
df.head()
```

```python
sns.histplot(df['Spending_USD'],kde = True)
```

```python
sns.histplot(df['Life_Expectancy'],kde = True,color = 'red')
```

```python
min_max = MinMaxScaler()
```

```python
data = min_max.fit_transform(df[['Spending_USD','Life_Expectancy']])
```

```python
data[:5]
```

```python
df[['Spending_USD','Life_Expectancy']] = pd.DataFrame(data)
```

```python
df.head()
```

```python
fig,axs = plt.subplots(1,2,figsize = (10,5))
sns.histplot(df['Spending_USD'],kde = True, ax = axs[0])
sns.histplot(df['Life_Expectancy'],kde = True,ax = axs[1],color = 'red')
plt.tight_layout()
plt.show()
```
