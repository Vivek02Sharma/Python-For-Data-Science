## Feature Scaling Standardization using scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.histplot(df.total_bill)
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill','tip']])
```

```python
data = scaler.transform(df[['total_bill','tip']])
```

```python
data
```

```python
df_scaling = pd.DataFrame(data,columns = ['total_bill','tip'])
```

```python
df_scaling.head()
```

```python
df[['total_bill','tip']] = df_scaling[['total_bill','tip']]
```

```python
df.head()
```

```python
sns.histplot(df.total_bill)
```
