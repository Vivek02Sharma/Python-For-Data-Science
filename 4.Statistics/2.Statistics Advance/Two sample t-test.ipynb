## Two sample t-test

##### What is Two sample t-test?
This test has another name as the independent samples t-test. It is basically used to check whether the unknown population means of given pair of groups are equal. tt allows one to test the null hypothesis that the means of two groups are equal.



```python
import scipy.stats as stats
import numpy as np

# Height of the students of two class
data_group1 = np.array([14, 15, 15, 16, 13, 8, 14,
                        17, 16, 14, 19, 20, 21, 15,
                        15, 16, 16, 13, 14, 12])
data_group2 = np.array([15, 17, 14, 17, 14, 8, 12,
                        19, 19, 14, 17, 22, 24, 16,
                        13, 16, 13, 18, 15, 13])
```

```python
np.mean(data_group1),np.mean(data_group2)
```

```python
np.var(data_group1),np.var(data_group2) #Here, the ratio is 12.260 / 7.7275 which is less than 4:1
```

```python
t_test,p_value = stats.ttest_ind(a = data_group1,b = data_group2,equal_var = True)
```

```python
t_test,p_value
```

```python
if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print('Fail to reject null hypothesis')
```
