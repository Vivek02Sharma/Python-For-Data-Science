## Covariance and Correlation

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = sn.load_dataset('titanic')
```

```python
df.head()
```

```python
#covariance 
df.cov()
```

```python
#correlation
df.corr()
```

```python
# pearson
df.corr(method = 'pearson')
```

```python
# spearman
df.corr(method = 'spearman')
```
