## Feature scaling Normalization

```python
import seaborn as sns
```

```python
df = sns.load_dataset('healthexp')
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max
```

```python
df.columns
```

```python
df.dtypes
```

```python
min_max.fit(df[['Year','Spending_USD','Life_Expectancy']])
```

```python
df1 = min_max.transform(df[['Year','Spending_USD','Life_Expectancy']])
```

```python
df1
```

```python
import pandas  as pd
```

```python
df1 =  pd.DataFrame(df1,columns = ['Year','Spending_USD','Life_Expectancy'])
```

```python
df1
```

```python
## we can combine the above step innstead one by one
df2 = min_max.fit_transform(df[['Year','Spending_USD','Life_Expectancy']])
```

```python
df2 = pd.DataFrame(df2,columns = ['Year','Spending_USD','Life_Expectancy'])
```
