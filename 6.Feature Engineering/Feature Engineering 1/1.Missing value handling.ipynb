### Handling missing value

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df[["age"]].mean()
```

```python
df['age'].median()
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df
```

```python
df['age'].isnull().sum()
```

```python
mode = df['embarked'].mode()[0]
```

```python
df['embarked'] = df['embarked'].fillna(mode)
```

```python
df.head()
```

```python
df.isnull().sum()
```
