## Data Interpolation

#### 1.Linear Interpolation

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,10,15,20,25])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.plot(x,y,"r--")
```

```python
## interplote the data using linear interpolation
x_new = np.linspace(1,5,100)
y_interp = np.interp(x_new,x,y)
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

 #### 2.Cubic interpolation using scipy

```python
x = np.array([1,2,3,4,5])
y = np.array(list(map(lambda x : x**3,x)))
```

```python
x,y
```

```python
plt.scatter(x,y)
```

```python
from scipy .interpolate import interp1d
```

```python
## create a cubic interpolation function
f = interp1d(x,y,'cubic')
```

```python
## interplate the data
x_new = np.linspace(1,5,100)
y_interp = f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

```python
x = np.array([1,2,3,4,5])
y = np.array(list(map(lambda x : x**3,x)))
```

```python
plt.scatter(x,y)
```

```python
## interpolate the data using polynomail interpolation
p = np.polyfit(x,y,4) # 4th degree
```

```python
x_new = np.linspace(1,5,30)
y_interp = np.polyval(p,x_new) # interpolate the y value
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```
