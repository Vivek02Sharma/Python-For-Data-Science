### SMOTE (Synthetic Minority Oversampling Technique)

```python
from sklearn.datasets import make_classification
```

```python
## creating imbalance data using classification technique
x,y = make_classification(n_samples=1000,
                    n_features=2,
                    n_clusters_per_class=2,
                    n_redundant=0,
                    weights=[0.80],
                    random_state=12
                    )
```

```python
x,y
```

```python
import pandas as pd
```

```python
data1 = pd.DataFrame(x,columns = ['f1','f2'])
data2 = pd.DataFrame(y,columns = ['target'])
```

```python
data1.head()
```

```python
data2.head()
```

```python
df1 = pd.concat([data1,data2],axis = 1)
```

```python
df1.head()
```

```python
df1['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df1['f1'],df1['f2'],c = df1['target'])
```

```python
## Over sampling using SMOTE method
from imblearn.over_sampling import SMOTE
```

```python
## performing over sampling 
oversample = SMOTE()
x,y = oversample.fit_resample(df1[['f1','f2']],df1[['target']])
```

```python
x.head()
```

```python
y.head()
```

```python
y['target'].value_counts()
```

```python
oversampled_df = pd.concat([x,y],axis = 1)
```

```python
oversampled_df.head()
```

```python
oversampled_df['target'].value_counts()
```

```python
plt.scatter(oversampled_df['f1'],oversampled_df['f2'],c = oversampled_df['target'])
```
