## Example 1: K-Fold Cross-Validation

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Generate synthetic dataset
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)

# Initialize KFold with k=5
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize a classifier (Logistic Regression in this case)
clf = LogisticRegression()

# Lists to store accuracy across folds
fold_accuracy = []

# Perform KFold cross-validation
for fold_idx, (train_index, test_index) in enumerate(kf.split(X), 1):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # Fit the model on the training data
    clf.fit(X_train, y_train)
    
    # Predict on the test data and compute accuracy
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    fold_accuracy.append(accuracy)
    
    # Print accuracy for each fold
    print(f"Fold {fold_idx}: Accuracy = {accuracy:.4f}")

# Plotting the accuracy across folds
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(fold_accuracy) + 1), fold_accuracy, marker='o', linestyle='-', color='b', label='Accuracy')
plt.title('Accuracy Across Folds in K-Fold Cross-Validation')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.xticks(np.arange(1, len(fold_accuracy) + 1))
plt.grid(True)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

## Example 2: Stratified K-Fold Cross-Validation

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Generate synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)

# Initialize Stratified K-Fold cross-validation with k=5
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize a classifier (Logistic Regression)
clf = LogisticRegression()

# Lists to store accuracy across folds
fold_accuracy = []

# Perform Stratified K-Fold cross-validation
for fold_idx, (train_index, test_index) in enumerate(skf.split(X, y), 1):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # Fit the model on the training data
    clf.fit(X_train, y_train)
    
    # Predict on the test data and compute accuracy
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    fold_accuracy.append(accuracy)
    
    # Print accuracy for each fold
    print(f"Fold {fold_idx}: Accuracy = {accuracy:.4f}")

# Plotting the accuracy across folds
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(fold_accuracy) + 1), fold_accuracy, marker='o', linestyle='-', color='b', label='Accuracy')
plt.title('Accuracy Across Folds in Stratified K-Fold Cross-Validation')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.xticks(np.arange(1, len(fold_accuracy) + 1))
plt.grid(True)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

## Example 3: Leave-One-Out Cross-Validation (LOOCV)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Generate synthetic dataset
X, y = make_classification(n_samples=100, n_features=5, n_classes=2, random_state=42)

# Initialize Leave-One-Out cross-validation
loo = LeaveOneOut()

# Initialize a classifier (Logistic Regression)
clf = LogisticRegression()

# Lists to store accuracy across iterations
accuracy_scores = []

# Perform Leave-One-Out cross-validation
for train_index, test_index in loo.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # Fit the model on the training data
    clf.fit(X_train, y_train)
    
    # Predict on the test data and compute accuracy
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Plotting the accuracy across iterations
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(accuracy_scores) + 1), accuracy_scores, marker='o', linestyle='-', color='b', label='Accuracy')
plt.title('Accuracy Across Iterations in Leave-One-Out Cross-Validation')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.xticks(np.arange(1, len(accuracy_scores) + 1))
plt.grid(True)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
