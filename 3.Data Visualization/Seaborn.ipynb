```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris.columns
```

```python
sns.scatterplot(x = 'sepal_length' , y = 'sepal_width' , data = iris )
```

```python
# also like these we can create scatterplot
sns.scatterplot(x = iris.sepal_length , y = iris.petal_length, hue = 'species',data = iris)
```

```python
sns.scatterplot(x = iris.sepal_length , y = iris.petal_length, hue = 'species',data = iris, style = 'species')
```

```python
sns.scatterplot(x = iris.petal_length, y = iris.petal_width , data = iris)
```

```python
iris.head()
```

```python
sns.scatterplot(x = iris.petal_length, y = iris.petal_width , data = iris, hue = 'species')
```

```python
sns.lineplot(x = iris.petal_length , y = iris.petal_width,hue = 'species' , style = 'species' , data = iris)
```

```python
sns.displot(x = iris.sepal_length)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip)
```

```python
tips['smoker'].value_counts()
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip,data = tips, hue = 'smoker',style = 'sex' )
```

```python
tips['size'].value_counts()
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip, data = tips, style = 'size', hue = 'time' )
```

```python
sns.catplot(x = tips.day, y = tips.total_bill,data = tips)
```

```python
sns.catplot(x = tips.smoker, y = tips.tip,data = tips)
```

```python
sns.jointplot(x = tips.total_bill, y = tips.tip)
```

```python
sns.jointplot(tips)
```
