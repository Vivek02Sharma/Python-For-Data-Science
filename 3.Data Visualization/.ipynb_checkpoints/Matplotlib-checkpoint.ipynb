## Matplotlib ##

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(1,10)
```

```python
y = np.sin(x)
```

```python
x = np.cos(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'--r')
plt.grid()
plt.figure(figsize = (8,8))
plt.xlabel("Random number")
plt.ylabel("sin")
plt.title("random data ")
plt.plot(x,y)
```

```python
a = np.random.rand(50)
b = np.random.rand(50)
```

```python
plt.scatter(a,b)
```

```python
color = np.random.rand(50)
size = 500 * np.random.rand(50)
plt.scatter(a,b, c = color ,s = size,alpha=0.5)
```

```python
plt.scatter(a,b, c = 'red' , s = size, alpha=0.8, marker = "x")
```

```python
data  = np.random.randint(1,10, size = 20)
data
```

```python
plt.hist(data)
plt.xlabel("number of data")
plt.ylabel("frequency of data")
plt.title("Histogram")
plt.show()
```

### Creating 3d plot ###

```python
x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)

fig = plt.figure()
ax = fig.add_subplot(projection = "3d")
color = np.random.randint(1,255,size = 20)
ax.scatter(x,y,z,c=color )
plt.xlabel("x data")
plt.ylabel("y data")
ax.set_zlabel("z data")
plt.show()
```
